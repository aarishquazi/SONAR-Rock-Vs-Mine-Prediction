```python
import pandas as pd
```

```python
df=pd.read_csv('sonar data.csv')
```

```python
ac={}
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
import sklearn
```

```python
y=df['R']
```

```python
y
```

```python
x=df.drop(columns='R',axis=1) 
```

```python
from sklearn import  model_selection
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
y_test
```

# Logistic Regression

```python
import sklearn.linear_model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
l=LogisticRegression()
```

```python
l.fit(x_train,y_train)
```

```python
y_pred=l.predict(x_test)
```

```python
y_test
```

```python
y_pred
```

```python
import sklearn.metrics
```

```python
from sklearn.metrics import accuracy_score
```

```python
ac['LogisticRegression']=accuracy_score(y_test,y_pred)*100
```

# Linear Support Vector Classifier

```python
from sklearn import svm
```

```python
svc=svm.LinearSVC()
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
ac['SVMC']=accuracy_score(y_test,y_pred)*100
```

# Gaussian Naive Bayes

```python
from sklearn import naive_bayes
```

```python
gnb=naive_bayes.GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
```

```python
ac['GaussianNB']=accuracy_score(y_test,y_pred)*100
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
from sklearn import tree
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
ac['Decision Tree']=accuracy_score(y_test,y_pred)*100
```

```python
tree.plot_tree(dt)
```

# KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
ac['KNeighborsClassifier']=accuracy_score(y_test,y_pred)*100
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
ac['RandomForestClassifier']=accuracy_score(y_test,y_pred)*100
```

# Accuracy Score of different models

```python
print("Accuracy score of different models :- \n")
for key,val in ac.items():
    print(key,val)
```
